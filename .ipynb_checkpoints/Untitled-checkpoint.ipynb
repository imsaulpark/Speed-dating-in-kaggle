```python
%matplotlib inline

import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt

sns.set(style="ticks")
np.random.seed(sum(map(ord, "axis_grids")))

data = pd.read_csv("data/Speed Dating Data.csv")
```

```python
data_1=data.dropna(subset=['match', 'age', 'age_o'])
g = sns.FacetGrid(data_1, col="match")
g.map(plt.scatter, "age", "age_o", alpha=.7)
g.add_legend()
```

```python
data_2=data.dropna(subset=['career_c','like'])
g = sns.FacetGrid(data_2, row="career_c", size=1.7, aspect=4,)
g.map(sns.distplot, "like", hist=False, rug=True);
```

```python
data_3=data.dropna(subset=['dec_o', 'gender', 'race'])
g = sns.FacetGrid(data_3, col="gender", size=4, aspect=.5)
g.map(sns.barplot, "race", "dec_o");
```

```python
import ast
data_4=data.dropna(subset=['match', 'income'])
for i,income in enumerate(data_4['income']):
    print(income)
    data_4['income'][i]=float(income.value())
g = sns.FacetGrid(data_4, row="match", size=1.7, aspect=4,)
g.map(sns.distplot, "income", hist=False, rug=True)
```
