# Data Import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("data/Speed Dating Data.csv")
data.head()
```

# Time1 (Before dating.)

## Data preprocessing

### 1. Missing value processing

```python
data['attr1_1'].fillna(0, inplace=True)
data['sinc1_1'].fillna(0, inplace=True)
data['intel1_1'].fillna(0, inplace=True)
data['fun1_1'].fillna(0, inplace=True)
data['amb1_1'].fillna(0, inplace=True)
data['shar1_1'].fillna(0, inplace=True)
```

### 2. Set dataframe for data in time1

```python
data_time1 = {'attractive': data['attr1_1'],
              'sincere' : data['sinc1_1'],
              'intelligent' : data['intel1_1'],
              'ambitious' : data['amb1_1'],
              'fun' : data['fun1_1'],
              'shared_interest' : data['shar1_1'] }
df_time1 = pd.DataFrame(data_time1)
df_time1.head()
```

### 3. EDC

```python
df_time1.describe()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3)
# sns.despine(left=True)

fig.set_size_inches(20,10)
plt.subplots_adjust(hspace=0.3)

axes[0, 0].set(title="Attractive")
axes[0, 0].set_xlim(0, 100)
sns.distplot(df_time1['attractive'], ax = axes[0, 0])

axes[0, 1].set(title="Sincere")
axes[0, 1].set_xlim(0, 100)
sns.distplot(df_time1['sincere'], ax = axes[0, 1])

axes[0, 2].set(title="Intelligent")
axes[0, 2].set_xlim(0, 100)
sns.distplot(df_time1['intelligent'], ax = axes[0, 2])

axes[1, 0].set(title="Fun")
axes[1, 0].set_xlim(0, 100)
sns.distplot(df_time1['fun'], ax = axes[1, 0])

axes[1, 1].set(title="Ambitious")
axes[1, 1].set_xlim(0, 100)
sns.distplot(df_time1['ambitious'], ax = axes[1, 1])

axes[1, 2].set(title="Shared Interest")
axes[1, 2].set_xlim(0, 100)
sns.distplot(df_time1['shared_interest'], ax = axes[1, 2])
```
