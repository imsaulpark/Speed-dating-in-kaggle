# Import Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import missingno as msno

df_data = pd.read_csv("data/Speed Dating Data.csv")
df_data.shape
```

# Set Dataframe by 'iid'
we define this dataframe as 'df_droppped'.
and we'll use this dataframe for making following dataframe about each personal property.
- profile
- time1, halfway, time2, time3

```python
df_dropped = df_data.drop_duplicates(["iid"])
df_dropped.set_index('iid', inplace=True)
df_dropped.head()
```

```python
# print missing values rate.
print(df_dropped.isnull().sum().sum() / df_dropped.size)

msno.matrix( df_dropped )
```

# 1. Profile

```python
df_profile = df_dropped.loc[ :, "age" : "expnum"]
df_profile.info()
```

```python
print( df_profile.shape )
# print missing values rate.
print( df_profile.isnull().sum().sum() / df_profile.size) 

msno.matrix( df_profile )
```
